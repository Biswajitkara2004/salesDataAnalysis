```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dft=pd.read_csv('/content/sales_data.csv', encoding='ISO-8859-1')
dft
```

```python
dft.info()
```

```python
#fetching null data from amount col-->
dft[dft['Amount'].isnull()]
```

```python
#Droping unwanted cols such as status,unnamed1
dft.drop('Status',axis=1,inplace=True)
```

```python
dft.drop('unnamed1',axis=1,inplace=True)
```

```python
dft.info()
```

```python
#Delete all the nan values from the dataframe
dft.dropna(inplace=True)
```

```python
dft.info()
```

```python
#rename th col Marital_status to marital
dft.rename(columns={'Marital_Status':'Marital'},inplace=True)
dft
```

```python
#change the values of marital col-->single,1-->married
dft['Marital'].replace({0:'single',1:'married'},inplace=True)
dft
```

```python
#change the values of Gender col F-->Female, M-->Male
dft['Gender'].replace({'F':'Female','M':'Male'},inplace=True)
dft
```

```python
#visualize the count of males and females based on gender column using seaborn
plt.figure(figsize=(20,20),facecolor='lightcyan')
plt.suptitle('~Sales DataAnalysis~',fontsize=30)
plt.subplot(3,3,1)
sns.countplot(x='Gender',data=dft)
plt.title('Gender Count')

a=sns.countplot(x='Gender',data=dft)
for i in a.containers:
  a.bar_label(i,fmt='%d')


plt.subplot(3,3,2)
gc=dft['Gender'].value_counts()
plt.pie(gc,labels=gc.index,autopct='%1.2f%%',explode=(0,0.1),shadow=True)
plt.title('Gender Pie Chart')


plt.subplot(3,3,3)
gs=dft.groupby('Gender',as_index=False)['Amount'].sum()
a= sns.barplot(x='Gender',y='Amount',data=gs,color='orange')
plt.title('Gender wise Amount')
for i in a.containers:
  a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightcyan')



plt.subplot(3,3,4)
gs=dft.groupby('Age Group',as_index=False)['Amount'].sum()
a= sns.barplot(x='Age Group',y='Amount',data=gs,color='orange')
plt.title('Gender wise Amount')
for i in a.containers:
  a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightcyan')


plt.subplot(3,3,5)
gs=dft.groupby('State',as_index=False)['Amount'].sum()
a= sns.barplot(y='State',x='Amount',data=gs,color='orange')
plt.xticks(rotation=90)
plt.title('Gender wise Amount')
for i in a.containers:
  a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightcyan')


plt.subplot(3,3,6)
gs=dft.groupby('Occupation',as_index=False)['Amount'].sum()
a= sns.barplot(y='Occupation',x='Amount',data=gs,color='orange')
plt.xticks(rotation=90)
plt.title('Gender wise Amount')
for i in a.containers:
  a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightcyan')

plt.subplot(3,3,7)
gs=dft.groupby('Product_Category',as_index=False)['Amount'].sum()
a= sns.barplot(y='Product_Category',x='Amount',data=gs,color='orange')
plt.xticks(rotation=90)
plt.title('Gender wise Amount')
for i in a.containers:
  a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightcyan')

plt.subplot(3,3,8)
gs=dft.groupby('Marital',as_index=False)['Amount'].sum()
a= sns.barplot(x='Marital',y='Amount',data=gs,color='orange')
plt.title('Gender wise Amount')
for i in a.containers:
  a.bar_label(i,fmt='%d')
plt.gca().set_facecolor('lightcyan')
plt.tight_layout()

plt.subplot(3,3,9)
gc=dft['Age Group'].value_counts()
plt.pie(gc,labels=gc.index,autopct='%1.2f%%',explode=(0,0,0,0.3,0,0,0),shadow=True)
plt.title('Gender Pie Chart')

plt.show()


```
